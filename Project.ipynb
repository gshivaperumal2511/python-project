```python
import pandas as pd
import numpy as np

# ---------- 1. LOAD CSV ----------
# change path if file location changes
file_path = "/content/sample_data/SampleSuperstore.csv"
data = pd.read_csv(file_path)

print("Dataset Loaded Successfully\n")

# ---------- 2. CONVERT TO NUMPY ARRAYS ----------
sales = data["Sales"].to_numpy()
quantity = data["Quantity"].to_numpy()
profit = data["Profit"].to_numpy()

# ---------- 3. TOTAL & AVERAGE SALES ----------
total_sales = np.sum(sales)
avg_sales = np.mean(sales)

print("Total Sales =", total_sales)
print("Average Sales =", avg_sales)

# ---------- 4. INDEXING & SLICING ----------
print("\nFirst 5 sales records:", sales[:5])
print("Last 5 quantity records:", quantity[-5:])

# ---------- 5. STATISTICAL OPERATIONS ----------
print("\nSales Statistics")
print("Mean =", np.mean(sales))
print("Median =", np.median(sales))
print("Variance =", np.var(sales))
print("Standard Deviation =", np.std(sales))

# ---------- 6. BROADCASTING ----------
# dataset has total Sales already, so estimate Unit Price
unit_price = sales / quantity   # vectorized division

# recompute revenue using broadcasting
revenue = quantity * unit_price

print("\nSample Computed Revenue:", revenue[:5])

# ---------- 7. VECTORIZED PROFIT ANALYSIS ----------
# profit margin %
profit_margin = (profit / sales) * 100

print("\nSample Profit Margin %:", profit_margin[:5])

# ---------- 8. CATEGORY-WISE EXAMPLE ----------
# total quantity sold (vectorized)
print("\nTotal Quantity Sold =", np.sum(quantity))
```
